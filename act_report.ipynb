- Insight 2 - **Favourites count vs Retweets count**

This insight tries to illustrate how favourites count(the number of likes a tweet gets when it is posted) and retweets count( the of times a tweet is reposted) are spread on the same scale.

```python
%matplotlib inline
import matplotlib. pyplot as plt
import matplotlib. image as mpimg
plt.figure(figsize=[12, 8])
img = mpimg. imread('Favs_retweets_counts.png')
imgplot = plt. imshow(img)
plt. show()
```

I decided to plot the histogram of favourite count and retweets count on the same plot, to compare the two distributions. We can observe that the distribution of favourite count and retweets count is similar, as small number of likes/retweets  have the highest frequency and big number of likes/retweets have the least frequency. This general shows a right skewed distributions. From the favourites count distribution, we can observe that the are quite a number of outliers compared to the retweets count distribution. We could trim the data to exclude favourites counts that are outliers.

- Insight 3 - **Dog breed analysis**

From the wrangled_act notebook, we saw that the dataset has quite a large number of dog breeds. I have decided to produce a visualisation that shows the most favoured dog breed in our dataset. The visualisation below shows the top 20 dog breeds with the most favourite counts.

```python
%matplotlib inline
import matplotlib. pyplot as plt
import matplotlib. image as mpimg
plt.figure(figsize=[18, 10])
img = mpimg. imread('Dog-breed-favs.png')
imgplot = plt. imshow(img)
plt. show()
```

I used the horizontal bar plot instead of a normal bar plot, so that the labels of each breed will be more visible. From the chart, we observe that the golden retriever is the most liked dog breed according to the WeRateDogs. The second most liked breed is the labrador retriever breed, with the Pembroke at the third position. We can say that the golden retriever is by far the most liked dog breed, this being illustrated by the length of the the horizontal bar. The basset breed is at position 20. The difference between the basset and golden retriever is quite large by just looking at the lengths of the two bars, where we can safely say that the golden retriever is the likeable breed.
